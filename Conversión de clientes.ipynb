De las descripciones obtenidas del dataframe, no parece haber problemas de tipos de datos. Sin embargo, hay columnas que se pueden eliminar porque su columna *count* de la descripción el valor es menor al 50% del total de los datos.

```python
for column in df.columns:
    if (df[column].aggregate('count') < 20027): #20.027 es el 50% de 40053
        df.drop(labels=column, axis=1, inplace=True)
df.shape[1]
```
